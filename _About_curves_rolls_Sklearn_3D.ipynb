## <center> ------------------ABOUT CURVES AND ROLLS IN 3D------------------ </center>

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
ax.view_init(7, -80)

from sklearn.datasets import make_swiss_roll
X, t = make_swiss_roll(n_samples=1000, noise=0.0,random_state=42)

ax.scatter(X[:, 0], X[:,1], X[:, 2], c=t, cmap='plasma')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
ax.view_init(7, -80)

from sklearn.datasets import make_s_curve
X, t = make_s_curve(n_samples=3000, noise=0.2,random_state=42)

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap='plasma')
plt.show()
```

```python
from numpy.random import default_rng
from numpy import sin, cos, pi, linspace
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time

rng, noise = default_rng(), 0.4
u, a = linspace(-4*pi, 4*pi, 300), 0.5

x = cos(u) + 2 * cos(u) * (1.5 + sin(1.5 * u / 2))
y = sin(u) + 2 * sin(u) * (1.5 + sin(1.5 * u / 2))
z = 2 * cos(1.5 * u)

x_nod, y_nod, z_nod = [], [], []
for i in range(300):
    for axis_nod_i, axis_i in zip([x_nod, y_nod, z_nod], [x, y, z]):
        axis_nod_i.append(axis_i[i] + rng.uniform(low=-noise, high=noise))
        
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z)
ax.scatter(x_nod, y_nod, z_nod)
plt.title('MOVED EAR IN 3D')
plt.show()
```
