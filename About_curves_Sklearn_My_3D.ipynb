## <center> ------------------MOVED CURVES IN 3D------------------ </center>

```python
from sklearn.datasets import make_s_curve
import matplotlib.pyplot as plt

def make_s_curve_3D(n_sampls=100, nois=0.0, random_stat=None):
    X1, t = make_s_curve(n_samples=n_sampls, noise = nois, random_state = random_stat)
    X1[:, 1] = X1[:, 1] + 2 * X1[:, 0] - 1
    X2 = X1.copy()
    X2[:, 2] = X2[:,2] + 1
    X2[:, 1] = -X2[:, 1]
    return X1, X2, t
```

```python
X1, X2, t = make_s_curve_3D(n_sampls=1000, nois=0.0, random_stat=None)

fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')

ax.view_init(5, 45)
ax.scatter(X1[:, 0], X1[:, 1] , X1[:, 2], c=t, cmap='plasma')
ax.scatter(X2[:, 0], X2[:, 1] , X2[:, 2], c=t, cmap='plasma')
ax.set_xlabel('X-axis', fontweight ='bold')
ax.set_ylabel('Y-axis', fontweight ='bold')
ax.set_zlabel('Z-axis', fontweight ='bold')
plt.show()
```
